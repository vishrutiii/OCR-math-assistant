```python
path = "./Images/sinx1.jpeg"
bodmas_img = "./Images/bodmas.jpeg"
```

```python
def read_image(path, rotate = True):
    img = cv2.imread(path)
    if rotate:
        img = np.rot90(img)
    img = cv2.resize(img, (600,600))
    plt.imshow(img)
    return img
```

```python
# filter
# thresholding
# morphological/ dilate
# detect edges
# contours, plot contours
```

```python
img = read_image(bodmas_img)
```

```python
# convert bgr image to hsv image to apply filters
hsv_image = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
# apply filter
```

```python
def conservative_smoothing_gray(data, filter_size):
    temp = []
    
    indexer = filter_size // 2
    
    new_image = data.copy()
    
    nrow, ncol = data.shape
    
    for i in range(nrow):
        
        for j in range(ncol):
            
            for k in range(i-indexer, i+indexer+1):
                
                for m in range(j-indexer, j+indexer+1):
                    
                    if (k > -1) and (k < nrow):
                        
                        if (m > -1) and (m < ncol):
                            
                            temp.append(data[k,m])
                            
            temp.remove(data[i,j])
            
            
            max_value = max(temp)
            
            min_value = min(temp)
            
            if data[i,j] > max_value:
                
                new_image[i,j] = max_value
            
            elif data[i,j] < min_value:
                
                new_image[i,j] = min_value
            
            temp =[]
    
    return new_image.copy()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
new_image = conservative_smoothing_gray(gray,5)
plt.figure(figsize=(11,6))
plt.subplot(121), plt.imshow(gray, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(new_image, cmap='gray'),plt.title('Conservative Smoothing')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
## Apply thresholding
thresh = cv2.adaptiveThreshold(new_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11, 22)
```

```python
plt.imshow(thresh, cmap = "gray")
```

```python
dilate = cv2.morphologyEx(thresh, cv2.MORPH_ERODE, np.ones((3,3)))
```

```python
plt.imshow(dilate, cmap = "gray")
```

```python
edge_image = cv2.Canny(dilate, 10, 150)
```

```python
plt.imshow(edge_image, cmap="gray")
plt.show()
```

```python
contours, hierarchy = cv2.findContours(dilate, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
dr_im = np.copy(img)
rectangle = []
for contour, hier in zip(contours[1:], hierarchy[0][1:]):
    area = cv2.contourArea(contour)
    # if area < 150:
    #     continue
    
    if area < 150 or hier[3] > 0: # Means Parent the Top Level Contour or Not
        continue
    x, y, w, h = cv2.boundingRect(contour)
    rectangle.append((x,y,w,h))
    cv2.rectangle(dr_im, (x, y), (x+w, y+h), (255,0,0), 1)

plt.imshow(dr_im)
plt.show()
```

```python
images = []
for i, rect in enumerate(rectangle):
    x, y, w, h = rect
    cv2.imwrite("./Expression/%d.jpg" % i, (thresh[y:y+h, x:x+w] != 255).astype("uint8")*255)
    im = (thresh[y:y+h, x:x+w] != 255).astype("uint8")
    images.append(im)
    plt.imshow(im, cmap="gray")
    plt.show()
```
